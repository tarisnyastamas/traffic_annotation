## Removing accents

```python
replacements = {
    "√°": "a",
    "√©": "e",
    "√≠": "i",
    "√≥": "o",
    "√∂": "o",
    "≈ë": "o",
    "√∫": "u",
    "≈±": "u",
    "√º": "u",
    "√≠": "i",
    "√≠": "i"
}

def clean_text_accents(text, replacements=replacements):
    text = "".join([replacements.get(c, c) for c in text])
    return text
```

```python
df['Corpus'] = [clean_text_accents(i) for i in df['Corpus']]
df
```

```python
with open('/home/gabor/Traffic annotation/hun2.json') as json_file:
    data_hu = json.load(json_file)
len(data_hu)
```

```python
cities_hu = {}
i = 0
for key, value in data_hu.items():
    print("---" + key + "---")
    
    for key2, value2 in value.items():
#         print("\t", key2, "->", value2)
        cities_hu[key2] = value2
        i += 1
print("processed %d items" % i)
```

```python
# create flat dict
coords_all = {}
i = 0
for key, value in data_ro.items():
    print("----" + key + "----")
    
    for key2, value2 in value.items():
#         print("\t", key2, value2)
        coords_all[key2] = value2
        i += 1
print("processed %d items" % i)
```

```python
final_dict = {}
i = 0
for key, value in cities_hu.items():
    if key in coords_all:
        print(key, "->", value)
        
        final_dict[key] = {
            'name_hun': value,
            'coords_x': coords_all[key][0],
            'coords_y': coords_all[key][1]
        }
        
        i += 1

print("Found %d matches" % i)
```

```python
with open('cities_coords.json', 'w') as outfile:
    json.dump(final_dict, outfile)
```
